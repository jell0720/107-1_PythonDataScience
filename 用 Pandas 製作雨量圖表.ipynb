```python
import pandas as pd

dflist = pd.read_html('https://www.cwb.gov.tw/V7/observe/rainfall/hb_100.htm')

filename = '雨量表.xlsx'
df = dflist[2]

df.columns = df.loc[0] #row 0 to column name 
df = df.drop(0).reset_index(drop=True) #drop row 0 and reset_index
df.to_excel(filename)

df = pd.read_excel(filename)

df.head(20)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

pd.read_excel(filename)

plt.rcParams['font.family']='DFKai-SB' #顯示中文 for Windows10
df[:15].plot(kind='bar',y='雨量(毫米)',x='測站名稱',figsize=(12,6),fontsize=18,title='雨量top15')
```
