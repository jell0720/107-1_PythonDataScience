## 定義 parseAddress() 函數，透過 google api 來分析行政區

parseAddress() 利用 google api 來分析行政區，優點是有些地址並未明確列出行政區，可用 google api 補齊。

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import requests
import json

def parseAddress(address):
    key = 'AIzaSyDMcWqgiq-8y3q7Vova_3qYi5zyqZM2Iec'
    response = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address='+address+'&language=zh-tw&key='+key)
    json = response.json() 
    for components in json['results'][0]['address_components']:
        if (components['types'][0]=='administrative_area_level_3'):
            return components['long_name']
        
df = pd.read_csv('商店街聯合發展協會會員.csv', engine='python', encoding='big5')
newcol = []
for address in df['地址']:
    print(address)
    tmp = parseAddress(address)
    print(tmp)
    newcol.append(tmp)

df['行政區'] = newcol
df.to_csv('商店街聯合發展協會會員_行政區.csv')

df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

df = pd.read_csv('商店街聯合發展協會會員_行政區.csv', engine='python', encoding='UTF-8')
se = df.groupby(by='行政區').size()
se = se.sort_values(ascending=False)
plt.rcParams['font.family']='DFKai-SB' #顯示中文 for Windows10
se.plot(kind='bar',y='數量',x='行政區',figsize=(12,6),fontsize=18,title='商店街聯合發展協會會員')

```
